### Line Plots
A line graph—also known as a line plot or a line chart—is a graph that uses lines to connect individual data points. A line graph displays quantitative values over a specified time interval. In finance, line graphs are commonly used to depict the historical price action of an asset or security.

Draw a line plot with possibility of several semantic groupings.

The relationship between x and y can be shown for different subsets of the data using the hue, size, and style parameters. These parameters control what visual semantics are used to identify the different subsets. It is possible to show up to three dimensions independently by using all three semantic types, but this style of plot can be hard to interpret and is often ineffective. Using redundant semantics (i.e. both hue and style for the same variable) can be helpful for making graphics more accessible.

## Bar Plots
Bar charts are one of the most comonly used graphical represehtations of data used to visually displays compare</br>
values to each other. They are easy to create ad interpret. They are also flexible and gave several variations of </br>
standard bar charts including vertical or horizontal bar charts, component or grouped chats, and stacker bar charts.

Data for a bar chart/plot are entered in columns. Each numeric  data value becomes a bar. The chart is constructed </br>
such that lengths of the different bars are propotional to the size of the category they represent X-axis represents </br>
the different categories and has no scale; the y-axis does have a scale and indicates the units of measurements, in case</br> of vertical bar charts, and vice versa in case of horizontal bar charts.

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
Phool = sns.load_dataset("iris")

# draw a line plot
sns.set_style("darkgrid")
sns.barplot(x="sepal_length", y="sepal_width" , data=Phool)
plt.title("Flowers Plot")

plt.show()
```

<h1 align='center'> Box-Plots </h1>
    
------------
The box plot is an exploratory graph. It is a standardized way of displaying the distribution of data based on       </br>
the five summary statistics minimum first quartile, median, third quartile, and maximum, First and third quartile    </br>
and maximum. First and third quartile is called two hinges, first quartile is the lower hinge and the third quartile </br> is the upper hinge. Minimum and the maximum are two whiskers. Minimum is the lower whisker and the maximum is the    </br> upper whisker. In other words we can say that box plot visualizes five summary statistics: the median,two hinges and </br> two whiskers.

In the simplest box plot the central triangle spans the first quartile to the third quartile (inter quartile range IQR).</br> A segment inside the rectangle shows the median and whiskers above and below the box show the locations of the      </br> minimum and maximum.

Box Plot is useful for identifying outliers and for comparing distributions. In other words we can say that box plot gives</br> us information about the location and variation in the data set. Particularly it helps us in detecting and ilustrating</br> location and variation changes between different groups of data.

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt
# canvas (Baloon board)

# load dataset
kashti = sns.load_dataset("titanic")

# draw a box plot
sns.set_style("darkgrid")
sns.boxplot(x="class", y="fare" ,data=kashti )
plt.title("Kashti Plot")
plt.show()
```

## 3- Dictionaries
- An unordered Collection of elements
- Key and Value
- Curly braces or brackets {}
- Mutateable/Change the values

```python
# Food and their Prices
food1= {"Mangoes":150, "Bananas":100, "Burger":90,"Dahi bally":50}
food1
```

```python
type(food1)
```

```python
# Extract data
keys=food1.keys()
keys
```
